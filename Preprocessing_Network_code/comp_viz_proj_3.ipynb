```python
file_path4 = "/Users/shreya/Desktop/Comp_viz/Proj_3/Data/Final_data.xls"
final_data = pd.read_excel(file_path4, sheet_name=0)
print(final_data.columns)
final_data
```

```python
final_data['Collab Impact'] = final_data['collab_songs'] / final_data['collab_individuals']
# Calculate the Engagement Score
# Update the 'Engagement_score' column directly
final_data.loc[:, 'Engagement_score'] = (
    0.5 * final_data['Likes'] +
    0.3 * final_data['Comments'] +
    0.2 * final_data['Views']
)

# Handle NaN or inf values, round to integers, and update in-place
final_data.loc[:, 'Engagement_score'] = (
    final_data['Engagement_score']
    .replace([np.inf, -np.inf], np.nan)  # Replace infinities with NaN
    .fillna(0)                          # Replace NaN with 0
    .round()                            # Round to the nearest integer
    .astype(int)                        # Convert to integer
)

# Display the updated DataFrame
final_data
```

```python
final_data = final_data.dropna()
final_data
```

```python
final_data.columns
```

```python
import numpy as np

# Calculate the Concert-to-View Ratio
final_data['Concert_View_Ratio'] = final_data['Est_Concert_Fee'] / final_data['Views']

# Calculate the average ratio across the dataset (equivalent to WINDOW_AVG)
window_avg = final_data['Concert_View_Ratio'].mean()

# Classify as "High", "Low", or "Medium" based on the average ratio
final_data['Fee_classification'] = np.where(
    final_data['Concert_View_Ratio'] > window_avg, 'High',
    np.where(final_data['Concert_View_Ratio'] < window_avg, 'Low', 'Medium')
)

# Display relevant columns to verify the result
print(final_data[['artist', 'Concert_View_Ratio', 'Fee_classification']])
final_data
```

```python
# Apply the logic to classify danceability
final_data['Danceability_classification'] = np.where(
    final_data['Danceability'] >= 0.8, 'Highly Danceable',
    np.where(final_data['Danceability'] >= 0.5, 'Moderately Danceable', 'Less Danceable')
)

# Display the updated DataFrame with the new classification column
print(final_data[['Danceability', 'Danceability_classification']])
```

```python
# Ensure no division by zero (replace 0 with NaN if needed)
final_data['interaction_sum'] = final_data['Views'] + final_data['Likes'] + final_data['Comments']

# Calculate the new column
final_data['concert_fee_to_interaction_ratio'] = (
    final_data['Est_Concert_Fee'] / final_data['interaction_sum']
) * 100

# Handle potential division by zero by replacing NaN or inf with 0
final_data['concert_fee_to_interaction_ratio'] = (
    final_data['concert_fee_to_interaction_ratio']
    .replace([np.inf, -np.inf], np.nan)
    .fillna(0)
)

# Display the relevant columns to verify
print(final_data[['Est_Concert_Fee', 'interaction_sum', 'concert_fee_to_interaction_ratio']])
```

```python
# Drop rows with any NaN values
final_data = final_data.dropna()
final_data
```

```python
# Rename 'concert_fee_to_interaction_ratio' to 'affordability_index'
final_data.rename(columns={'concert_fee_to_interaction_ratio': 'affordability_index'}, inplace=True)
final_data
```

```python
# Min-Max scaling to scale 'Engagement_score' to range [1, 100]
min_score = final_data['Engagement_score'].min()
max_score = final_data['Engagement_score'].max()

final_data['Engagement_score'] = (
    (final_data['Engagement_score'] - min_score) / (max_score - min_score) * 99 + 1
).round().astype(int)  # Scale to [1, 100] and convert to integer

# Min-Max scaling to scale 'Engagement_score' to range [1, 100]
min_score = final_data['affordability_index'].min()
max_score = final_data['affordability_index'].max()

final_data['affordability_index'] =10 -( (
    (final_data['affordability_index'] - min_score) / (max_score - min_score) * 9 + 1)
).round().astype(int)  # Scale to [1, 100] and convert to integer
```
