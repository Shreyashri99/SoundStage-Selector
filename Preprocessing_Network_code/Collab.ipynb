```python
import pandas as pd
import numpy as np
```

```python
file_path = "/Users/shreya/Desktop/Comp_viz/Proj_3/Data/Final_data.csv" 
final_data = pd.read_csv(file_path)

print(final_data.columns)
final_data
```

```python
# Drop the unnecessary "Unnamed" columns
final_data = final_data.loc[:, ~final_data.columns.str.contains('^Unnamed')]

# Group by 'artist', count 'Track', sum 'Views', 'Likes', and 'Comments', and take the first value of other fields
final_data_grouped = final_data.groupby('artist').agg({
    'Track': 'count',                # Count of tracks
    'collab_individuals': 'first',   # Take the first value of collab_individuals
    'collab_songs': 'first',
    'country': 'first',              # Take the first value of country
    'Views': 'sum',                  # Sum of views
    'Likes': 'sum',                  # Sum of likes
    'Comments': 'sum',               # Sum of comments
    'Avg_Concert_Fee': 'first'       # Take the first value of Avg_Concert_Fee
}).reset_index()

# Rename 'Track' count column for clarity
final_data_grouped.rename(columns={'Track': 'Track Count'}, inplace=True)

# Calculate Engagement Score
final_data_grouped['Engagement_score'] = (
    0.5 * final_data_grouped['Likes'] +
    0.3 * final_data_grouped['Comments'] +
    0.2 * final_data_grouped['Views']
).round().astype(int)
```

```python
final_data_grouped
```

```python
# Min-Max scaling to scale 'Engagement_score' to range [1, 10]
min_score = final_data_grouped['Engagement_score'].min()
max_score = final_data_grouped['Engagement_score'].max()

final_data_grouped['Engagement_score_scaled'] = (
    (final_data_grouped['Engagement_score'] - min_score) / (max_score - min_score) * 99 + 1
).round(0)  # Scale to [1, 10] and round to 2 decimal places
```

```python
final_data_grouped
```

```python
final_data_grouped['Collab Impact'] = final_data_grouped['collab_songs'] / final_data_grouped['collab_individuals']
```

```python
min_score = final_data_grouped['Collab Impact'].min()
max_score = final_data_grouped['Collab Impact'].max()

final_data_grouped['Collab Impact'] = (
    (final_data_grouped['Collab Impact'] - min_score) / (max_score - min_score) * 9 + 1
).round(2)  # Scale to [1, 10] and round to 2 decimal places
```
